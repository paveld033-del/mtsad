## 2.3. Выявление и обработка выбросов
Для поиска выбросов использовались:
- межквартильный размах;
- boxplot;
- z-оценки / другой метод.

Для анализа выбросов применялись методы IQR и z-score.

````python
def iqr_outlier_summary(dataframe, cols):
    results = []

    for col in cols:
        series = dataframe[col].dropna()
        if len(series) == 0:
            continue

        q1 = series.quantile(0.25)
        q3 = series.quantile(0.75)
        iqr = q3 - q1

        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr

        outliers = series[(series < lower) | (series > upper)]

        results.append({
            "column": col,
            "q1": q1,
            "q3": q3,
            "iqr": iqr,
            "lower_bound": lower,
            "upper_bound": upper,
            "outlier_count": outliers.shape[0],
            "outlier_percent": round(100 * outliers.shape[0] / len(series), 2)
        })

    return pd.DataFrame(results).sort_values("outlier_percent", ascending=False)


important_numeric = [col for col in [
    "Train Speed",
    "Maximum Speed",
    "Gross Tonnage",
    "Equipment Damage Cost",
    "Track Damage Cost",
    "Total Damage Cost",
    "Total Persons Injured",
    "Total Persons Killed",
    "Temperature"
] if col in df.columns]

iqr_summary = iqr_outlier_summary(df, important_numeric)
display(iqr_summary)
````

````python
from scipy.stats import zscore

def zscore_outlier_summary(dataframe, cols, threshold=3):
    results = []

    for col in cols:
        series = dataframe[col].dropna()
        if len(series) < 3:
            continue

        zscores = np.abs(zscore(series))
        outlier_count = (zscores > threshold).sum()

        results.append({
            "column": col,
            "outlier_count_zscore": int(outlier_count),
            "outlier_percent_zscore": round(100 * outlier_count / len(series), 2)
        })

    return pd.DataFrame(results).sort_values("outlier_percent_zscore", ascending=False)

zscore_summary = zscore_outlier_summary(df, important_numeric)
display(zscore_summary)
````

### Результаты:

Наибольшее количество выбросов наблюдается в:

* `Total Damage Cost` — **13.28%**
* `Equipment Damage Cost` — **12.68%**
* `Track Damage Cost` — **12.56%**
* `Train Speed` — **12.08%**

При использовании z-score:

* выбросы составляют примерно **1–2%** наблюдений

### Принятое решение:

* выбросы **не удалялись автоматически**, поскольку:

  * большие значения ущерба — реальные аварии;
  * редкие события — важны для анализа риска;
* для моделей могут использоваться:

  * логарифмирование;
  * масштабирование (StandardScaler / MinMaxScaler).

**Обоснование:**
В задачах анализа аварий экстремальные значения являются важной частью данных, а не ошибкой.
